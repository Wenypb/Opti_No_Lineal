# Función objetivo

```python
import numpy as np
def funcion(x):
  return 2*(x[0]**2)-2*(x[0]*x[1])+9*(x[1]**2)+11*(x[0])-9*(x[1])+6
```

# Gradiente de la función objetivo

```python
def gradiente(x):
  g = np.empty((2,1))
  g[0] = 4*x[0]-2*x[1]+11
  g[1] = -2*x[0]+18*x[1]-9
  return g
```

# Método de reducción del tamaño de paso

```python
#Vector inicial
xk=np.array([[-5],[3]])
i=0
a=1
Tol=0.0001
gradk=np.array(gradiente(xk))
nor=np.linalg.norm(gradk)
x0s=[]
while nor > Tol:
  x0=xk
  fk_1=funcion(x0)
  xk=x0-a*gradk
  fk=funcion(xk)
  #Cuenta las iteracciones
  i=i+1
  x0s.append(xk)
  while fk_1<fk:
    a=0.5*a
    fk_1=funcion(x0)
    xk=x0-a*gradk
    fk=funcion(xk)
  gradk=np.array(gradiente(xk))
  nor=np.linalg.norm(gradk)
sol=xk
print("Con un total de",i,"iteracciones se llego a este vector solucion:")
print(sol)
```

# Grafica del método de reducción de tamaño de paso

```python
import matplotlib.pyplot as plt
x = np.arange(-60,60)
y = np.arange(-60,60)

w= np.meshgrid(x,y)  # malla
Z = funcion(w)
arreglo = np.array(x0s).reshape((i,-1))
plt.style.use('seaborn-v0_8-poster')
plt.contour(w[0],w[1],Z, levels=20)
plt.plot(sol[0],sol[1],'ro')
plt.plot(arreglo[:,0],arreglo[:,1],'orange')  # iteraciones del método
plt.xlim([-60,60])
plt.ylim([-60,60])
plt.show()
```
