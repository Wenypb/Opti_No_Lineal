```python
import numpy as np

def funcion(x):
  return 2*(x[0]**2)-2*(x[0]*x[1])+9*(x[1]**2)+11*(x[0])-9*(x[1])+6

a=2
b=-2
c=9
d=11
e=-9
f=6
#MATRIZ A
A = np.array([[a, b/2], [b/2, c]])
#VECTOR B
B = np.array([[d],[e]])
#C
C=np.array([f])
#Vector inicial
x0=np.array([[-5],[3]])
#Gradiente x0
g= (2*A)@x0+B
Tol=0.0001
nor=np.linalg.norm(g)
i=0
x0s=[]
while nor>Tol:
  g= (2*A)@x0+B
  pk=np.array(-g)
  pkt=pk.T
  ak= -(pkt@g)/(pkt@(2*A)@pk)
  xk=x0-ak*g
  x0s.append(xk)
  x0=xk
  nor=np.linalg.norm(g)
  #Conteo de iteracciones
  i=i+1
print("El vector solucion es")
print(x0)
print("Con",i,"iteracciones")

import matplotlib.pyplot as plt
x = np.arange(-15,15)
y = np.arange(-5,5)

w= np.meshgrid(x,y)  # malla
Z = funcion(w)
arreglo = np.array(x0s).reshape((i,-1))
plt.style.use('seaborn-v0_8-poster')
plt.contour(w[0],w[1],Z, levels=20)
plt.plot(x0[0],x0[1],'ro')
plt.plot(arreglo[:,0],arreglo[:,1],'orange')  # iteraciones del m√©todo
plt.xlim([-15,15])
plt.ylim([-5,5])
plt.show()
```
